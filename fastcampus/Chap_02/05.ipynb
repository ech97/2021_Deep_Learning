```python
import os
from glob import glob

import tensorflow as tf 
import numpy as np

import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data_paths = glob('../../../dataset/cifar/train/*.png')
path = data_paths[0]
path
```

```python
gfile = tf.io.read_file(path)
image = tf.io.decode_image(gfile)
image.shape
```

```python
plt.title(os.path.basename(path)) # 이름만 쏘옥
plt.imshow(image)
plt.show()
```

```python
def read_image(path):
    gfile = tf.io.read_file(path)
    image = tf.io.decode_image(gfile)
    return image
```

```python
dataset = tf.data.Dataset.from_tensor_slices(data_paths) #data path가 들어감
dataset = dataset.map(read_image) # path들을 image들로 대응시켜 치환 
dataset
```
