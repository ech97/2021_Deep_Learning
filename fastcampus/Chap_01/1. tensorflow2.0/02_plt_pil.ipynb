## Basic Attributes

alpha: 투명도

kind: 그래프 종류 'line', 'bar', 'barh', 'kde'

logy: y축 log로 스케일링

rot: 눈금 이름 회전시키기(0~360)

xticks, yticks: x, y축으로 사용할 값 (xlabel))

xlim, ylim: x, y축 범위

grid: 격자 on/off

figzise(,): 창 크기

sharex, sharey: 서브플롯들의 x,y축을 공유하고 눈금ㄱ

````python
from PIL import Image

import numpy as np 
import matplotlib.pyplot as plt
import cv2
````

````python
path1 = 'img/01.jpg'
path2 = 'img/02.jpg'

# 흑백
image_1 = Image.open(path1).convert("L") #Gray Scale (RGB채널에서 한개의 채널로 바뀜. 다만 0~256 수치를 색으로") #Gray Scale (RGB채널에서 한개의 채널로 바뀜. 다만 0~256 수치를 색으로 표현한것뿐))
image_bw = np.array(image_1)
print(image_bw.shape)

image_2 = Image.open(path2).convert("L")
image_bw2 = np.array(image_2)
image_bw2 = cv2.resize(image_bw2, (259, 194))





# 두개의 이미지 사이즈 맞추기
plt.subplot(1, 2, 1)
plt.imshow(image_bw, 'RdBu')
plt.colorbar()
plt.show()


plt.subplot(1, 2, 2)
plt.imshow(image_bw2, 'jet')
plt.colorbar()
plt.show()






````
